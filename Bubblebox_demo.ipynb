# Tilstander

<table style="margin:0px;width:100%;height:100%;">
<tr style="background-color:rgba(0, 0, 0, 0);width:100%">
    <td>
<font size="3" face="Courier New" align = "center" >
    <p>Et <i>system</i> er en samling molekyler som kan vekselvirke (dytte og trekke på hverandre) under visse betingelser. Disse betingelsene kan være gitt ved <i>randbetingelser</i> altså hvilke regler som gjelder når molekylene når randen av systemet, og det kan være <i>initialbetingelser</i>, altså hvordan molekylene er definert ved tiden 0.</p>
    <p>Et system har - i likhet med et enkeltmolekyl - et sett <i>koordinater</i> som definerer systemets tilstand. Der hvor et monatomisk molekyl i tre dimensjoner som regel kan defineres ved 6 koordinater; 3 for posisjon og 3 for hastighet, kan systemet som helhet defineres ved $6N$ koordinater, hvor $N$ er antall molekyler i systemet. (Vi ser her bort fra elektronenes frihetsgrader).</p>
    <p>Et gitt sett med koordinater svarer til en såkalt <i>mikrotilstand</i>, hvor vi kjenner alt som er å vite om systemet. Generelt finner vi at mange ulike mikrotilstander kan resultere i de samme målingene på makroskopisk nivå, slik som for eksempel temperatur, og vi sier derfor at alle slike like mikrotilstander svarer til den samme <i>makrotilstanden</i>. En makrotilstand er bestemt for et isolert system ved energi, volum og antall partikler. </p>
<p>I et realistisk laboratorieforsøk er antallet mikrotilstander som regel for stort til å kunne bestemmes, og det viser seg dessuten at vi ikke behøver å vite eksakt hvilken mikrotilstand systemet befinner seg i for å forutsi hvordan systemet utvikler seg. Det er derfor de fleste ligningene du benytter i KJM1130 ikke inneholder et utall partikkelposisjoner, men snarere makroskopiske variable som trykk og temperatur. I et numerisk forsøk, slik vi nå skal sette opp, vil vi derimot kjenne alle detaljer om systemet vi modellerer, hvilket gir oss en unik mulighet til å forstå sammenhengen mellom mikro- og makrotilstander. </p>
    <p>
    La oss til å begynne med sette opp et system med BubbleBox og visualisere mikrotilstanden det befinner seg i:
    </p>

# Systemet
### En todimensjonal Lennard-Jones væske

<div style="margin:0px">
<table style="margin:0px;width:100%;height:100%;">
<tr style="background-color:rgba(0, 0, 0, 0);width:100%">
    <td>
<font size="3" face="Courier New" >
    <p></p>
<p>Kreftene som virker mellom molekyler i reelle kjemiske systemer skyldes elektrostatiske og kvantemekaniske interaksjoner mellom elektroner og protoner. I vårt tilfelle vil vi basere oss på den generelle observasjonen at mange molekyler trekkes mot hverandre på lange avstander, men støtes vekk fra hverandre på korte avstander. I akkurat passe avstand kan de derfor i blant inngå bindinger og holde sammen i par eller små grupper, og det er også mulig for molekyler å inngå i større gitterstrukturer slik som i en krystall.</p
    </td>
</tr>
</table>
<table style="margin:0px;width:100%;height:100%;">
        <tr style="background-color:rgba(0, 0, 0, 0);width:50%;">
            <td style="text-align: left; vertical-align: upper;"><font size="3" face="Courier New" align = "center" >
<p>Vi kan modellere all denne oppførselen med en enkel modell hvor interaksjonen mellom molekylene er beskrevet med <i>Lennard-Jones krefter</i>, gitt ved uttrykket</p>
<p></p>
\begin{equation}
\mathbf{F}_{ij} = \epsilon \Big{(} \big{(}\frac{\sigma}{r_{ij}} \big{)}^6 - \big{(}\frac{\sigma}{r_{ij}} \big{)}^{12} \Big{)} \frac{\mathbf{r_{ij}}}{ r_{ij}^2 }. \quad \quad (1)
\end{equation}
                <p></p>
                <p>Her er $\mathbf{r}_{ij} = \mathbf{r}_j - \mathbf{r}_i$ vektoren mellom molekylene $i$ og $j$, med avstanden $r_{ij}$, og parameterne $\sigma$ og $\epsilon$ gis passende tallverdier. Inntil videre kan du anta at $\sigma = \epsilon = 1$, noe som forenkler uttrykket betydelig. </p>
                <p>I figuren til høyre kan du se hvordan kraften mellom to molekyler virker frastøtende for små verdier av $r_{ij}$, og tiltrekkende for større avstander. </p>

    <p>
Kraften som virker på et enkelt molekyl i systemet vil være summen av alle kreftene fra de andre molekylene. Matematisk kan vi uttrykke dette som
    <p></p>
        \begin{equation}
\mathbf{F}_{i} = \sum_{j \neq i} \mathbf{F}_{ij}. \quad \quad (2)
\end{equation}
    </p>
    <p>
    Det er nyttig for deg å vite dette, for det forklarer også hvorfor slike modeller kan være kostbare å regne på; for hver inkrementell endring av tid i simuleringen må vi regne kreftene for hvert par av molekyler, summere dem sammen og fra dette bestemme akselerasjonen til det enkelte molekyl. 
    </p>
        <p>Når vi velger at molekylene skal vekselvirke på denne måten, og vi i tillegg begrenser molekylenes bevegelse til 2 dimensjoner ved å la posisjoner, hastighet og krefter beskrives av todimensjonale vektorer, vil systemet vi modellerer konvensjonelt kalles en <i> todimensjonal Lennard-Jones væske</i>. Du kan lese mer om slike modeller <a href="her">her</a>.</p>

# Likevekt
<table style="margin:0px;width:100%;height:100%;">
<tr style="background-color:rgba(0, 0, 0, 0);width:100%">
    <td>
<font size="3" face="Courier New" align = "center" >
<p>Du har nå sett hvordan du lager en todimensjonal Lennard-Jones væske med BubbleBox og hvordan du lar den uvikle seg i tid. Du spør deg kanskje; <i>hva ønsker vi å oppnå med å la dette systemet utvikle seg</i>? <i>Når er det i mål</i>? </p>
    <p>Det korte svaret er at vi er i mål når det <i>ikke lenger skjer noe interessant</i> i systemet. På samme måte som i et laboratorieforsøk må vi definere et kriterium for når reaksjonen er over, og når det er tid for å analysere resultatene. 
    </p>
<p>På laboraoriumet vil du muligens gjøre observasjoner av opphør i tempereratur- eller faseendringer, og bruke disse som indikasjon på at reaksjonen er over. I numeriske forsøk, slik du skal utføre i BubbleBox, kan du benytte tilsvarende observasjoner, for eksempel opphør av endring i kinetisk energi. Disse observasjonene forteller deg at systemet har oppnådd <i>likevekt</i>, altså at det befinner seg i den mest sannsynlige makrotilstanden. </p>
    <p>La oss derfor se nærmere på hvordan den kinetiske energien i systemet varierer over tid. Vi trenger først å kunne regne denne energien, som for et molekyl $i$ er uttrykt ved:</p>
    <p></p>
    \begin{equation}
    E_{K,i} = \frac{1}{2} m_i v_i^2. \quad \quad (3)
    \end{equation}
    <p></p>
    <p>Her er $m_i$ massen til molekyl $i$ og $v_i^2 = v_{x,i}^2 +   v_{y,i}^2$ (i to dimensjoner). For systemet som helhet er den kinetiske energien derfor</p>
    <p></p>
    \begin{equation}
    E_{K} = \sum_i^{N} E_{K,i}, \quad \quad (4)
    \end{equation}
    <p></p>
    <p>hvor N er antall molekyler. For å regne denne trenger vi altså å vite massen og hastigheten til molekylene, og i BubbleBox er disse til enhver tid tilgjengelig i numpy-arrays som:</p>
    <p></p>
</td>
</tr>
</table>

````python
def kinetic_energy_readable(system):
    # Funksjon med hensyn på leselighet
    
    n_molecules = system.vel.shape[1] #antall molekyler
    
    E_k = 0 # kinetisk energi
    
    for i in range(n_molecules): #summer sammen enkeltenergier
        m_i = system.masses[i] # massen til molekyl i
        v_ix = system.vel[0,i] # x-komponenten til hastigheten til molekyl i
        v_iy = system.vel[1,i] # y-komponenten til hastigheten til molekyl i
    
        E_k = E_k + 0.5 * m_i *(v_ix**2 + v_iy**2) 
    
    return E_k


def kinetic_energy_fast(system):
    # Vektorisert funksjon med hensyn på ytelse
    return .5*bb.np.sum(system.masses*bb.np.sum(system.vel**2, axis = 0))


print("Total kinetisk energi (leselig): %.2f" % kinetic_energy_readable(s))
print("Total kinetisk energi ( kjapp ): %.2f" % kinetic_energy_fast(s))
````
